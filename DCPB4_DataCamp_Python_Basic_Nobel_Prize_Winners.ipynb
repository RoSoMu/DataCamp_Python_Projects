The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
nobel = pd.read_csv("data/nobel.csv")
print(nobel.head())
print(nobel.info())
print(nobel.shape)

# Most commonly awarded gender:
nobel_gender = nobel.groupby("sex")["laureate_id"].agg("count")
nobel_gender = nobel_gender.reset_index()
nobel_gender_srt= nobel_gender.sort_values("laureate_id", ascending = False)
print(nobel_gender_srt)
top_gender = str(nobel_gender_srt.iloc[0, 0])
print(top_gender)

# Most commonly awarded country:
nobel_country = nobel.groupby("birth_country")["laureate_id"].agg("count")
nobel_country = nobel_country.reset_index()
nobel_country_srt= nobel_country.sort_values("laureate_id", ascending = False)
print(nobel_country_srt.head())
top_country = str(nobel_country_srt.iloc[0, 0])
print(top_country)

# Decade with the highest ratio of US-born Nobel Prize winners:
total_winners = 1000
nobel_us_born = nobel[nobel["birth_country"] == "United States of America"]
nobel_us_born["decade"] = nobel_us_born["year"]//10*10
nobel_us_born_grouped= nobel_us_born.groupby("decade").size().reset_index()
nobel_us_born_grouped["ratio"] = nobel_us_born_grouped[0]/1000
nobel_us_born_srtd = nobel_us_born_grouped.sort_values("ratio", ascending = False)
max_decade_usa = int(nobel_us_born_srtd.iloc[0,0])
print(max_decade_usa)

# Decade and Nobel Prize Category combination with the highest proportion of female winners:
nobel["female_winner"] = nobel["sex"] == "Female"
nobel["decade"] = nobel["year"]//10*10
prop_female_winner = nobel.groupby(["decade", "category"], as_index = False)["female_winner"].mean()
prop_female_winner_srt = prop_female_winner.sort_values("female_winner", ascending = False)
print(prop_female_winner_srt.head())
decade = prop_female_winner_srt.iloc[0,0]
print(decade)
category = prop_female_winner_srt.iloc[0,1]
print(category)
max_female_dict = {decade: category}
print(max_female_dict)

                   
# First woman to receive a Nobel Prize and category:
nobel_female= nobel[nobel["sex"] == "Female"].sort_values("year")
nobel_female_sliced = nobel_female[["year", "full_name", "category"]]
print(nobel_female_sliced.head())
first_woman_name = str(nobel_female_sliced.iloc[0,1])
print(first_woman_name)
first_woman_category = str(nobel_female_sliced.iloc[0,2])
print(first_woman_category)

# Repeat winners:
repeat_winners = nobel.groupby("full_name").size().reset_index().sort_values(0, ascending = False)
repeat_winners_all = repeat_winners[repeat_winners[0] > 1]
print(repeat_winners_all)
repeat_list = list(repeat_winners_all["full_name"])
print(repeat_list)
```
